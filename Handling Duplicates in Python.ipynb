```python
import pandas as pd
import numpy as np
```

```python
week_name = ['Week 1','Week 1','Week 1','Week 2','Week 2','Week 2','Week 3','Week 4','Week 4','Week 4']
city =  ['Jaipur', 'Bhopal','Indore','Gurugram','Jaipur','Jaipur','Bhopal','Gurugram','Indore','Indore']
metric = [20,40,50,90,20,20,30,10,40,10]
data = pd.DataFrame(list(zip(week_name,city,metric)), columns = ['Week','City','Metric'])
```
